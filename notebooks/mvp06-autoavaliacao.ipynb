# Autoavaliação do MVP

## Cumprimento dos objetivos

O objetivo deste MVP foi desenvolver um pipeline completo em nuvem capaz de transformar dados brutos de prontuário sintético (Synthea) em um Data Warehouse modelado em esquema estrela e, a partir dele, gerar indicadores hospitalares analíticos.

Esse objetivo foi atingido por meio de um pipeline em camadas (Bronze, Silver e Gold), desde a ingestão e padronização dos dados até a consolidação das tabelas fato e dimensões na camada Gold. O modelo dimensional resultante permitiu calcular e explorar, de forma reprodutível, os três indicadores centrais do projeto: mortalidade intra-hospitalar, tempo de permanência (LOS) e readmissão em até 30 dias, com segmentações por tempo, diagnóstico clínico e perfil demográfico (idade/faixa etária, sexo e raça).

As perguntas de negócio definidas no início do trabalho foram respondidas diretamente a partir das tabelas da camada Gold, e o notebook de análises demonstrou consultas representativas sem necessidade de transformações adicionais fora do modelo dimensional, reforçando a adequação do esquema estrela para o objetivo do MVP. Ressalta-se que, por se tratar de dados sintéticos, os valores obtidos têm finalidade demonstrativa (validação do pipeline e do modelo analítico), não epidemiológica.

## Dificuldades encontradas

As principais dificuldades estiveram relacionadas à padronização e integração dos dados clínicos do Synthea. Apesar de sintéticos, os dados não refletem integralmente convenções comuns em prontuários do contexto brasileiro, como a codificação explícita por CID (Classificação Internacional de Doenças) e a distinção estruturada entre diagnósticos primários e secundários. Além disso, alguns desfechos e eventos não são representados de forma tão explícita quanto em sistemas reais, exigindo derivação de métricas a partir de regras (ex.: campos estruturados para motivo de desfecho da internaçao como óbito ou alta melhorado).

Também foram identificados registros artefatuais e variações temporais fora do período de interesse, o que demandou aplicação de filtros e regras adicionais de qualidade para garantir consistência e plausibilidade das análises.

Por fim, foi necessário calibrar o nível de detalhamento do modelo para equilibrar realismo e simplicidade: o escopo foi concentrado nas tabelas essenciais e métricas prioritárias para viabilizar uma entrega completa, rastreável e alinhada às perguntas de negócio.

## Trabalhos futuros

Como continuidade, podem ser implementadas extensões em duas frentes:

- Evolução do modelo: inclusão de novas dimensões clínicas (procedimentos, medicamentos, comorbidades).
- Evolução analítica: ampliação do conjunto de indicadores, estratificações adicionais e integração com ferramentas de visualização interativa, além da aplicação do pipeline a dados reais anonimizados, respeitando requisitos éticos e de privacidade.

Essas extensões aproximariam o MVP de um cenário de produção e reforçariam seu valor como demonstração de engenharia de dados aplicada à saúde.
